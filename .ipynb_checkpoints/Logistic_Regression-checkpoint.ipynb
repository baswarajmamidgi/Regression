```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
def sigmoid(scores):
    return 1 / (1 + np.exp(-scores))
```

```python
def log_likelihood(X, y, weights):
    scores = np.dot(X, weights)
    ll = np.sum( y*scores - np.log(1 + np.exp(scores)) )
    return ll
```

```python
#loading breast_cancer data set

breast_cancer = datasets.load_breast_cancer()
X = breast_cancer.data  # we only take the first two features.
y = breast_cancer.target


plt.scatter(X[:, 0], X[:, 1],
            c = y, alpha = .4)
plt.show()
```

```python
def logistic_regression(X, y):
   
    weights = np.zeros(X.shape[1])
    num_steps = 100000
    learning_rate = 5e-5
    
    for step in range(num_steps):
        scores = np.dot(X, weights)
        predictions = sigmoid(scores)

        # Update weights with gradient
        output_error_signal = y - predictions
        gradient = np.dot(X.T, output_error_signal)
        weights += learning_rate * gradient
        
        # Print log-likelihood every so often
        #if step % 20000 == 0:
            #print (log_likelihood(X, y, weights))
        
    return weights
```

```python
weights = logistic_regression(X, y)
```

```python
#prediction
final_scores = np.dot(10, weights)
preds = np.round(sigmoid(final_scores))
preds
```
